# Problem 1
### Given 2 strings (string_a and sting_b) check whether or not anagrams, given following criteria:
> Strings only anagrams if all conditions below are met:
> * They must be exactly same length
> * They must use exactly same characters
Example: cars and scar, heart and earth.

```python
# if len(string_a) != len(string_b):
#     return False
# if string_a length does equal string_b length
# sort characters
# if sorted characters in string_a and string_b are not exact match
#      return False
# else if all match return True

def is_anagram(string_a, string_b):                   # function declaration
    if len(string_a) != len(string_b):                # checking string length match
        return False                                  # guarantees best case time complexity = O(1)
    char_times_a = dict()                             # 
    char_times_b = dict()                             # declaring two dictionaries to track characters
    for i in range(len(string_a)):                    # looping for each value in range 1 to len(str)
        if string_a[i] not in char_times_a.keys():    # checking if we've seen char
            char_times_a[string_a[i]] = 0             # if not, record in dictionary
        else:                                         #
            char_times_a[string_a[i]] += 1            # if so, increment how many times seen it
        if string_b[i] not in char_times_b.keys():    
            char_times_b[string_b[i]] = 0
        else:
            char_times_b[string_b[i]] += 1
    return char_times_a == char_times_b               # check if dictionaries are equal by end of loop
        
```

```python
is_anagram("cars", "scar")
```

```python
class Stack:
    def __init__(self):
        self.items = []

    def is_empty(self):
        return self.items == []

    def push(self, item):
        self.items.append(item)
    
    def pop(self):
        return self.items.pop()

    def peek(self):
        return self.items[len(self.items)-1]

    def size(self):
        return len(self.items)

    def __len__(self):
        return self.size()
```

# Problem 2: Reverse a string
### use stack to reverse a string


def reverse_str(my_string):
    stack = Stack()
    for char in my_string:
        stack.push(char)
    string_out = ""
    while stack.is_empty() == False:
        string_out += stack.pop()
    return string_out
        

```python
reverse_str('blah blah blah')
```

```python
reverse_str('')
```

```python
class Queue:
    def __init__ (self):
        self.items = []
    
    def is_empty(self):
        return self.items == []

    def enqueue(self, item):
        self.items.insert(0, item)
    
    def dequeue(self):
        return self.items.pop()

    def size(self):
        return len(self.items)
```
